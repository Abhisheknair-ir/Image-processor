```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

# Function to scrape images
def scrape_images(url):
    # Create a directory to save images
    os.makedirs('images', exist_ok=True)

    # Get the content of the webpage
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all image tags
    img_tags = soup.find_all('img')

    # Loop through all image tags and download the images
    for img in img_tags:
        img_url = img.get('src')
        if not img_url:
            continue
        # Join the URL if it's relative
        img_url = urljoin(url, img_url)

        try:
            # Get the image response
            img_response = requests.get(img_url)
            img_response.raise_for_status()  # Check for request errors

            # Write the image to a file
            img_name = os.path.join('images', os.path.basename(img_url))
            with open(img_name, 'wb') as f:
                f.write(img_response.content)

            print(f'Downloaded: {img_name}')
        except Exception as e:
            print(f'Failed to download {img_url}: {e}')

# Example usage
if __name__ == "__main__":
    website_url = 'https://www.ajio.com/s/min-70-percent-off-5582-59581?query=%3Arelevance%3Agenderfilter%3AMen%3Al1l3nestedcategory%3AMen%20-%20Shirts%3Al1l3nestedcategory%3AMen%20-%20Trousers%20%26%20Pants&curated=true&curatedid=min-70-percent-off-5582-59581&gridColumns=3&segmentIds='  # Replace with the target website
    scrape_images(website_url)
```

```python
import os
from PIL import Image
from rembg import remove

# Define input and output directories
INPUT_DIR = 'D:/TAG/V01/Trial_DataSet/Model01/Input'
OUTPUT_DIR = 'D:/TAG/V01/Trial_DataSet/Model01/Output'

os.makedirs(OUTPUT_DIR, exist_ok=True)

def process_image(image_path):
    # Open the image
    with open(image_path, 'rb') as input_file:
        input_image = input_file.read()
    
    # Remove background
    output_image = remove(input_image)
    
    # Save the output image
    output_path = os.path.join(OUTPUT_DIR, f'processed_{os.path.basename(image_path)}')
    with open(output_path, 'wb') as output_file:
        output_file.write(output_image)
    
    return output_path

def main():
    # Loop through all files in the input directory
    for filename in os.listdir(INPUT_DIR):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            original_path = os.path.join(INPUT_DIR, filename)
            processed_path = process_image(original_path)
            print(f'Processed and saved: {processed_path}')

if __name__ == '__main__':
    main()
```
